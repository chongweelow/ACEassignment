### Reggresion plot

```python
# Visualization of dataset (Regression model).
fig, subs   = plt.subplots(1, 3, figsize = (30,10))

sns.regplot(data = df, x = 'Min_Temp', y = 'FFB_Yield', ax = subs[0], color = 'tab:red'); 
sns.regplot(data = df, x = 'Max_Temp', y = 'FFB_Yield', ax = subs[1], color = 'tab:green'); 
sns.regplot(data = df, x = 'Average_Temp', y = 'FFB_Yield', ax = subs[2], color = 'tab:blue'); 

subs[0].title.set_text('Min_Temp vs FFB_Yield')
subs[1].title.set_text('Max_Temp	 vs FFB_Yield')
subs[2].title.set_text('Average_Temp vs FFB_Yield')

 
```

```python
# Visualization of dataset (Regression model).
fig, subs   = plt.subplots(1, 3, figsize = (30,10))

sns.regplot(data = df, x = 'SoilMoisture', y = 'FFB_Yield', ax = subs[0], color = 'tab:red'); 
sns.regplot(data = df, x = 'Precipitation', y = 'FFB_Yield', ax = subs[1], color = 'tab:green'); 
sns.regplot(data = df, x = 'HA_Harvested', y = 'FFB_Yield', ax = subs[2], color = 'tab:blue'); 

subs[0].title.set_text('SoilMoisture vs FFB_Yield')
subs[1].title.set_text('Precipitation vs FFB_Yield')
subs[2].title.set_text('HA_Harvested vs FFB_Yield')

 
```

```python
# Visualization of dataset (Regression model).
fig, subs   = plt.subplots(1, 3, figsize = (30,10))


sns.regplot(data = df, x = 'Precipitation', y = 'SoilMoisture', ax = subs[0], color = 'tab:red'); 
sns.regplot(data = df, x = 'Average_Temp', y = 'SoilMoisture', ax = subs[1], color = 'tab:green'); 
sns.regplot(data = df, x = 'Max_Temp', y = 'SoilMoisture', ax = subs[2], color = 'tab:blue');  

subs[0].title.set_text('Precipitation vs SoilMoisture')
subs[1].title.set_text('Average_Temp vs SoilMoisture')
subs[2].title.set_text('Max_Temp vs SoilMoisture')

 
```

Findings :
---
----------------
1. Precipitation is positively associated with FFB_Yield.
2. HA_Harvested is negatively associated with FFB_Yield.
3. SoiMoisture feature show no significant association with FFB_Yield dierecly.However Precipition is positively associated with SoiMoisture.
4. Temperature feature show no significant association with FFB_Yield directly. However Averge_Temp & Max_Temp is negatively associated with SoiMoisture.




Conclusions:
---
----------------------
Based on the above findings feature like Precipitation, HA_Harvested and SoiMoisture are important factors influence fresh fruit bunch (FFB) yield. FFB Yield is positively affected by Precipitation and SoiMoisture is influenced by Precipitation positively. Hence, the increment of the SoiMoisture need to be considered by monitoring the tempature to increase FFB Yield.
