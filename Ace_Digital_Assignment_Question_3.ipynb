```python
import urllib.request
```

```python

file = 'https://raw.githubusercontent.com/chongweelow/ACEassignment/main/Data.txt'
document = urllib.request.urlopen(file).read().decode('UTF-8')
```

```python
import re
def tokenize(string):
    return re.compile('\w+').findall(string)

from collections import Counter

def word_freq(string): 
    text = tokenize(string.lower())
    c = Counter(text)           # count the words
    d = Counter(''.join(text))  # count all letters
    return (dict(c),dict(d))    # return a tuple of counted words and letters

data = document
words, letters = word_freq(data) # count and get dicts with counts

sumWords = sum(words.values())       # sum total words
sumLetters = sum(letters.values())   # sum total letters

# calc / print probability of word
for w in words:
    print("Probability of '{}': {}".format(w,words[w]/sumWords))

# update the counts to propabilities:
for w in words: 
    words[w] = words[w]/sumWords

print ( words) 
```

```python
# stopword removal
file = 'https://raw.githubusercontent.com/chongweelow/ACEassignment/main/stopwords.txt'
```

```python
import urllib.request 
stopwords = urllib.request.urlopen(file).read().decode('UTF-8').splitlines()
```

```python
cleaned_tokens = [token for token in tokens if token not in stopwords]
print(cleaned_tokens)
```

```python
import string
translation = str.maketrans("", "", string.punctuation)
cleaned_tokens = [token.translate(translation) for token in cleaned_tokens if token not in stopwords]
```

```python
print(Counter(cleaned_tokens).most_common(30))
```

```python
cleaned_tokens = [token for token in cleaned_tokens if len(token) > 0]
print(Counter(cleaned_tokens).most_common(30))
```

```python
import pandas as pd
df = pd.DataFrame(Counter(cleaned_tokens).most_common(30), columns=['term', 'frequency'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='term', y='frequency',data=df);
plt.xticks(rotation=90);
```
